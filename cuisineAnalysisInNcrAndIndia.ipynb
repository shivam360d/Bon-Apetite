### Pie Charts show the top 10 cuisines across NCR region and Rest of India.
>1. Irrespective of the region, $North Indian$ and $Chinese$ cuisines are the top candidate.<br>
>2. There is <ins>no significant difference</ins> between the <ins>number of cuisines</ins> served in NCR and Rest of India region.<br>
>3. $Bakery$ and $Street$ food cuisines comes in top10 list of NCR and are not in rest of India.<br>
>4. $Mexican$ and $Cafe$ cuisines comes in top10 list of rest of India and are not in NCR.

#### Cuisine Analysis:
>Comparision evident in Pie charts shown below.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/home/shivam/Downloads/zomato.csv',encoding='latin-1')
data.dropna(inplace=True)
```

```python
India = data[data['Country Code']==1]
```

```python
NCR = {}
Rest = {}
India['Cuisines']=India['Cuisines'].str.strip().str.split(",")
for i in range(India.shape[0]):
    for j in range(len(India.Cuisines.iloc[i])):
        key=India.Cuisines.iloc[i][j].strip()
        if India.iloc[i]['City'] in ['New Delhi','Gurgaon','Noida','Faridabad','Ghaziabad']:
            NCR[key] = NCR.get(key,0) + 1
        else:
            Rest[key] = Rest.get(key,0) + 1
```

```python
NCR = sorted(NCR.items(), key = lambda kv:(kv[1], kv[0]),reverse=True)
Rest = sorted(Rest.items(), key = lambda kv:(kv[1], kv[0]),reverse=True)
```

```python
cuisinesNCR = []
countNCR = []
for i in range(10):
    cuisinesNCR.append(NCR[i][0])
    countNCR.append(NCR[i][1])
```

```python
cuisinesRest = []
countRest = []
for i in range(10):
    cuisinesRest.append(Rest[i][0])
    countRest.append(Rest[i][1])
```

```python
plt.pie(countNCR,labels=cuisinesNCR,autopct="%.1f%%")
plt.title('distribution of top 10 cuisines in NCR')
plt.show()
```

```python
plt.pie(countRest,labels=cuisinesRest,autopct="%.1f%%")
plt.title('distribution of top 10 cuisines in Rest of India')
plt.show()
```

```python
# cuisines not common in top 10 list:
print(set(cuisinesNCR) ^ set(cuisinesRest))
```
