```python
import numpy as np
import pandas as pd
import geopandas as gpd
```

```python
data = pd.read_csv('data_sampled.csv',index_col=0)
```

```python
ratings = data.groupby('State')['Aggregate rating'].mean()
```

```python
ratings = pd.DataFrame(ratings)
```

```python
geo_data = gpd.read_file("Indian_States.json")
```

```python
geo_data = geo_data.merge(ratings,how="left",left_on="NAME_1",right_on="State")
```

```python
geo_data["Aggregate rating"].fillna(0,inplace=True)
```

```python
plt = geo_data.plot(column="Aggregate rating",legend=True,edgecolor='black',figsize=(8,15),cmap="Reds",scheme="quantiles")
```
