```python
from geopy.geocoders import Nominatim
# initialize Nominatim API
geolocator = Nominatim(user_agent="shivam.vats@colorado.edu")
```

```python
data['Longitude'] = data['Longitude'].astype("str") 
data['Latitude'] = data['Latitude'].astype("str")
data['lat/lon'] = data['Latitude'].str.cat(data['Longitude'],sep=',')
```

```python
def get_state(position):
    location = geolocator.reverse(position)
    return str(location).split(",")[-3].strip()
```

```python
data_sampled = data.groupby('City').apply(lambda x: x.sample(20) if x.shape[0] >= 20 else x.sample(x.shape[0])).reset_index(drop=True)
```

```python
data_sampled['City'].value_counts()
```

```python
data_sampled["State"] = data_sampled["lat/lon"].apply(get_state)
```

```python
data_sampled.to_csv("data_sampled.csv")
```
